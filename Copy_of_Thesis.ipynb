4-USING PYSPAN FOR CLEANING DATA


* CHECKING FOR UNIQUE VALUES IN THE DATA


```python
data.nunique()
```

* COUNTING ALL THE NULL VALUES IN THE DATA

---




```python
null_count = data.isnull().sum()
print(null_count)
```

* REMOVING THE NULL VALUES FROM THE DATA


```python
data_cleaned = ps.handle_nulls(data, action='remove')
null_count = data_cleaned.isnull().sum()
print(null_count)
```

* CHECKING FOR DUBLICATES IN THE DATA





```python
duplicates = data.duplicated()
print(duplicates)

```

* CHECKING FOR OUTLIERS IT IN THE DATA & REMOVING IT.





```python
new_data = ps.detect_outliers(data_cleaned, columns = ['Charging Duration (hours)','Temperature (°C)'], method = 'iqr')
```

* SPLITTING THE EXISTING COLUMN OF CHARGING START TIME & CHARGING END TIME'
BASED ON THE SPACES AND CREATED NEW ONE


```python
df = pd.DataFrame(data)
df_split = df['Charging Start Time'].str.split(' ', expand=True)
df_split2 =df['Charging End Time'].str.split(' ', expand=True)
```


* RENAMING THE NEW COLUMNS


```python
df_split.columns = ['Start Charging Date', 'Start Charging Time']
df_split2.columns = ['End Charging Date', 'End Charging Time']
print(df_split)
print(df_split2)
```

* REMOVING  EXISTING COLUMNS



```python
new_data.drop(columns=['Charging End Time', 'Charging Start Time'], inplace=True)
```

* ADDING THE NEW COLUMN TO EXISTING DATA


* CONVERTING THE DATE COLUMNS TO DATETIME FORMAT


```python
df_split = pd.DataFrame(df_split)
df_split2 = pd.DataFrame(df_split2)
df_split['Start Charging Date'] = pd.to_datetime(df_split['Start Charging Date'])
df_split2['End Charging Date'] = pd.to_datetime(df_split2['End Charging Date'])
```

* SORTING EACH DATEFRAME BY ITS DATE COLUMN


```python
df_split_sorted =  df_split.sort_values(by='Start Charging Date').reset_index(drop=True)
df_split2_sorted = df_split2.sort_values(by='End Charging Date').reset_index(drop=True)
```

* ADDING THE NEW  SORTED COLUMNS TO EXISTING DATA


```python
new_data['Start Charging Date'] = df_split_sorted['Start Charging Date']
new_data['Start Charging Time'] = df_split_sorted['Start Charging Time']
new_data['End Charging Date'] = df_split2_sorted['End Charging Date']
new_data['End Charging Time'] = df_split2_sorted['End Charging Time']
new_data.columns
```

**STARTING WITH DATA VISUALIZATION**

* **CHECKING THE PERCENTAGE OF EACH USER TYPE IN DATA**

```python
# Set the figure size to 8 inches by 8 inches
plt.figure(figsize=(8, 8))

# Count the repeatation of each user type in the 'User Type' column of new_data
user_types = new_data['User Type'].value_counts()

# Set the values for x-axis and y-axis
bar_xaxis = user_types.index
bar_yaxis = user_types.values

# Create an explode array to separate each slice of the pie chart
explode = [0.1] * len(bar_xaxis)

# Display the percentage of each slice with one decimal place, explode the slices, add a shadow, and use the 'viridis' color palette
plt.pie(bar_yaxis, labels=bar_xaxis, autopct='%1.1f%%', explode=explode, shadow=True, colors=sns.color_palette("viridis", len(bar_xaxis)))

# Create labels for x-axis & y-axis
plt.title("Types of User", fontsize=20, fontweight='bold')
plt.xlabel("User Type", fontsize=15)
```

*  **ALANLYZING THE TOP 5 CHARGING STATION**

```python
# Set the figure size to 12 inches by 6 inches
plt.figure(figsize=(12, 6))

# Count the  top 5 repeatation of each Charging Station Location in the new_data
cs_breakdown = new_data['Charging Station Location'].value_counts().head()

# Set the values for x-axis and y-axis
cs_xaxis = cs_breakdown.index
cs_yaxis = cs_breakdown

# Set the style of the seaborn plot to 'viridis'
sns.barplot(x=cs_xaxis, y=cs_yaxis, palette='viridis', dodge=False)

# Add titles and labels
plt.title("Top 5 Charging Stations", fontsize=20, fontweight='bold')
plt.ylabel("Load", fontsize=15)
plt.xlabel("Charging Station Names", fontsize=15)
plt.xticks(rotation=45, ha='right')

# Add value labels on top of the bars
for index, value in enumerate(cs_yaxis):
    plt.text(index, value + 0.5, str(value), ha='center', fontsize=12)
```
